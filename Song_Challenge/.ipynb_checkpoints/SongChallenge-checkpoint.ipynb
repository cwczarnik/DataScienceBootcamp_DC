```python
## a group by will give me a list likethis.
data.sort_values('user_sign_up_date',ascending = False)
df_to_rank = data.groupby(['user_state','user_sign_up_date','user_id']).sum().reset_index()
```

```python
def ranker(df):
    """Assigns a rank to each employee based on salary, with 1 being the highest paid.
    Assumes the data is DESC sorted."""
    df['date_rank'] = np.arange(len(df)) + 1
    return df

ranked_df = df_to_rank.sort_values('user_sign_up_date').groupby('user_state').apply(ranker)
```

```python
So we get all the states and their range
ranked_df[ranked_df['date_rank'] ==1].shape[0]/len(set(data['user_state']))
```

## Build a function that takes as an input any of the songs in the data and returns the most likely song to be listened next. That is, if, for instance, a user is currently listening to "Eight Days A Week", which song has the highest probability of being played right after it by the same user? This is going to be v1 of a song recommendation model.

```python
# import sklearn metrics for distances, we will use the cosine similarity metrics
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.metrics.pairwise import manhattan_distances
from sklearn.metrics.pairwise import euclidean_distances
```

```python
data.head()
```

```python
data_2 = pd.get_dummies(data,columns =['song_played'],prefix='',prefix_sep='')
data_2 = data_2.groupby('user_id').sum().reset_index()
# data_2 = data_2[data_2['user_id'] == 4]
X =data_2[data_2.columns.difference(['id','time_played','user_id','user_sign_up_date','user_state','counts','date','hour'])]
X = X.T
X
```

```python
dists = cosine_similarity(X,X)
dists_named = pd.DataFrame(dists,columns = X.index)
dists_named.index = dists_named.columns
```

```python
dists_named.head()
```

```python
## Smaller the eucilidean distance the more "similar" the song
```

```python
song_name = 'Across The Universe'
p = dists_named[song_name].sort_values(ascending = False)
mask = p.index.isin([song_name])
p = p[~mask]
p[p>0][0:5]
```

Cosine similarity metric finds the normalized dot product of the two attributes. By determining the cosine similarity, we would effectively try to find the cosine of the angle between the two objects. The cosine of 0° is 1, and it is less than 1 for any other angle.

It is thus a judgement of orientation and not magnitude: two vectors with the same orientation have a cosine similarity of 1, two vectors at 90° have a similarity of 0, and two vectors diametrically opposed have a similarity of -1, independent of their magnitude.
